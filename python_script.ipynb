```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
#deleting the last column
df.drop('2023', axis=1)
#axis=0 -->x axis(row), axis=1 -->y axis(column)
#drop for deleting
```

```python
#taking only population data
df_new=df.iloc[:,4:67]
```

```python
df_new
```

```python
df_new.columns
```

```python
for i in df_new.columns:
    fig=plt.figure(figsize=(5,5))
    plt.hist(df_new[i],color='#022222',bins=10)
    plt.xlabel(i)
    plt.show()
```

```python
years=df_new.columns[1:]
total_values= df_new[years].sum()
plt.figure(figsize=(30,30))
plt.barh(years,total_values,color='#191970')
plt.xlabel('total values')
plt.ylabel('year', size=20)
plt.title('Total values per year',size=20)
plt.show()
```
